```python
from sklearn.preprocessing import MinMaxScaler

MMS = MinMaxScaler()
df_high = MMS.fit_transform(df_high.values.reshape(-1,1))
df_high
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
df_high = scaler.fit_transform(np.array(df_high).reshape(-1,1))
df_high
```

```python
train_size = int(len(df_high) * 0.8)
test_size = len(df_high) - train_size

train_size, test_size
```

```python
train_data, test_data = df_high[:train_size, :], df_high[train_size:,:1]
train_data.shape
```

```python
# Previous 100 data points are stored
def create_data(data, time_step):
    x_data, y_data = [], []
    for i in range(len(data) - time_step -1):
        prev = data[i: i+time_step, 0]
        x_data.append(prev)
        y_data.append(data[i+time_step , 0])
        return np.array(x_data), np.array(y_data)
```

```python
time_step = 100
x_train, y_train = create_data(train_data, time_step)
x_test, y_test = create_data(test_data, time_step)
```

```python
# Make 3.0 data(LSTM Required) (samples, time steps, features)
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1],1)
x_train.shape, y_train.shape
```

**MODEL BUILDING**

```python
#Import necessary Libraries
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import tensorflow as tf
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.layers import LSTM
%matplotlib inline
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences = True, input_shape = (100,1)))
model.add(LSTM(50, return_sequences = True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')


model.summary()
```

```python
model.fit(x_train, y_train, validation_data = (x_test, y_test), epochs = 100, batch_size = 64, verbose = 1)
```

```python

train_pre = model.predict(x_train)
test_pre = model.predict(x_test)
```

```python
train_pre = scaler.inverse_transform(train_pre)
test_pre = scaler.inverse_transform(test_pre)
```

```python
math.sqrt(mean_squared_error(y_train, train_pre))
```
